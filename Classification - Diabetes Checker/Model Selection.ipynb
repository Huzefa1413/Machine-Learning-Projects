# Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix,roc_auc_score,roc_curve

from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression

from sklearn.ensemble import BaggingClassifier
```

# Splitting Data into Training Set and Testing Set

```python
dataset = pd.read_csv("diabetes.csv")
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
dataset
```

## Applying KNN 

```python
knn = KNeighborsClassifier()
knn.fit(X_train,y_train)
knn_pred = knn.predict(X_test)
accuracy_knn = accuracy_score(y_test, knn_pred)
precision_knn = precision_score(y_test, knn_pred)
recall_knn = recall_score(y_test, knn_pred)
f1_knn= f1_score(y_test, knn_pred)
knn_confusion = confusion_matrix(y_test, knn_pred)
knn_auc_score = roc_auc_score(y_test, knn_pred)

print("KNN Accuracy:", "{:.3f}".format(accuracy_knn))
print("KNN Precision:", "{:.3f}".format(precision_knn))
print("KNN Recall:", "{:.3f}".format(recall_knn))
print("KNN F1 score:", "{:.3f}".format(f1_knn))
print("KNN AUC Score:", knn_auc_score)

plt.figure(figsize=(5, 4))
sns.heatmap(knn_confusion, annot=True, cmap='Blues', fmt='g', xticklabels=['Predicted No Diabeties', 'Predicted Diabeties'], yticklabels=['Actual No Diabeties', 'Actual Diabeties'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()

fpr, tpr, thresholds = roc_curve(y_test, knn_pred)
plt.plot(fpr, tpr, label='ROC Curve (AUC = {:.2f})'.format(knn_auc_score))
plt.plot([0, 1], [0, 1], linestyle='--', color='gray', label='Random')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.show()
```

```python
nb = GaussianNB()
nb.fit(X_train,y_train)
nb_pred = nb.predict(X_test)
accuracy_nb = accuracy_score(y_test, nb_pred)
precision_nb = precision_score(y_test, nb_pred)
recall_nb = recall_score(y_test, nb_pred)
f1_nb= f1_score(y_test, nb_pred)
nb_confusion = confusion_matrix(y_test, nb_pred)
nb_auc_score = roc_auc_score(y_test, nb_pred)

print("Naive Bayes Accuracy:", "{:.3f}".format(accuracy_nb))
print("Naive Bayes Precision:", "{:.3f}".format(precision_nb))
print("Naive Bayes Recall:", "{:.3f}".format(recall_nb))
print("Naive Bayes F1 score:", "{:.3f}".format(f1_nb))
print("Naive Bayes AUC Score:", nb_auc_score)

plt.figure(figsize=(5, 4))
sns.heatmap(nb_confusion, annot=True, cmap='Blues', fmt='g', xticklabels=['Predicted No Diabeties', 'Predicted Diabeties'], yticklabels=['Actual No Diabeties', 'Actual Diabeties'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()

fpr, tpr, thresholds = roc_curve(y_test, nb_pred)
plt.plot(fpr, tpr, label='ROC Curve (AUC = {:.2f})'.format(nb_auc_score))
plt.plot([0, 1], [0, 1], linestyle='--', color='gray', label='Random')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.show()
```

# Evaluating the performance of all models

```python
models = ['KNN', 'Naive Bayes', 'Decision Tree', 'Random Forest', 'Logistic Regression', 'Bagging']

accuracies = [accuracy_knn, accuracy_nb, accuracy_dt, accuracy_rf, accuracy_lr, accuracy_bg]
precisions = [precision_knn, precision_nb, precision_dt, precision_rf, precision_lr, precision_bg]
recalls = [recall_knn, recall_nb, recall_dt, recall_rf, recall_lr, recall_bg]
f1_scores = [f1_knn, f1_nb, f1_dt, f1_rf, f1_lr, f1_bg]
auc_scores = [knn_auc_score, nb_auc_score, dt_auc_score, rf_auc_score, lr_auc_score, bg_auc_score]

metrics_data = {'Accuracy': accuracies, 'Precision': precisions, 'Recall': recalls, 'F1 Score': f1_scores, 'AUC Score': auc_scores}

# Plotting
fig, axs = plt.subplots(3, 2, figsize=(10, 12))

for i, (metric, data) in enumerate(metrics_data.items()):
    row = i // 2
    col = i % 2
    ax = axs[row, col]
    ax.bar(models, data, color='blue')
    ax.set_title(f'{metric} of Different Models')
    ax.set_ylabel(metric)
    ax.set_ylim(0, 1)
    ax.tick_params(axis='x', rotation=45)
    for idx, val in enumerate(data):
        ax.text(idx, val + 0.01, f'{val:.2f}', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
